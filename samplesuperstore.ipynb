```python
df.dropna(inplace=True)
```

```python
df['Sales']=df['Sales'].round(0).astype(int)
```

```python
df.info()
```

```python
df['Profit']=df['Profit'].round(0).astype(int)
```

```python
df['Discount']=df['Discount'].round(0).astype(int)
```

```python
df['Quantity']=df['Quantity'].round(0).astype(int)
```

```python
df.columns
```

```python
ax= sns.countplot(x='Category',data=df,color='green') 
for bars in ax.containers:
    ax.bar_label(bars)


```

```python
from the above graph we can see that among the three product category  office supplies category is highest on sales.
```

```python
top10_subcat=df.groupby('Sub-Category')['Sales'].sum().sort_values(ascending=False).head(10)
top10_subcat
```

```python
plt.figure(figsize=(8,5))
sns.barplot(x=top10_subcat.values,y=top10_subcat.index)
plt.title("Top 10 sub-categories by sales")
plt.xlabel("Total Sales")
plt.ylabel("Sub-Category")
plt.show()
```

```python
phones is the highest selling sub category,making it the top revenue contributor among  all sub categories.this indicates strong
customer demand for phone related products.
```

```python
df.columns
```

```python
top_10_State=df.groupby('State')['Sales'].sum().sort_values(ascending=False).head(10).reset_index()
top_10_State
```

```python
plt.figure(figsize=(12,6))
sns.barplot(x='Sales',y='State',data=top_10_State,palette='coolwarm')
plt.title('top 10 State by Sales',fontsize=16)
plt.xlabel('Total Sales',fontsize=12)
plt.ylabel('State',fontsize=12)
plt.show()
```

```python
from above graph we can see that California has the highest total sales ,contributing significantly more than other states. Also the lowest state in 
terms of total sales is virginia.
```

```python
df.columns
```

```python
Sales_State=df.groupby(['State'],as_index=False)['Quantity'].sum().sort_values(by='Quantity',ascending=False).head(10)

sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data=Sales_State,x='State',color='pink',y='Quantity')
```

```python
from above graph we can see that most of the quantity are being sold to california. and the last in terms of quantity is michigan.
```

```python
top_10_City=df.groupby("City")["Profit"].sum().sort_values(ascending=False).head(10)
top_10_City
```

```python
import matplotlib.pyplot as plt
top_10_City.plot(kind="bar",x="City",y="Profit",color='red'),
plt.title("top 10 City by Profit")
plt.xticks(rotation=45)
plt.show()
```

```python
in the above graph it is visible that the most profit generating city is new york city which means the pricing is done effectively and efficiently here 
focusing on profit maximisation. where as on the other hands San Diego is the least profit genaerating city.
```

```python
df.columns
```

```python
Region_Sales=df.groupby("Region")["Sales"].sum().sort_values(ascending=False)
plt.figure(figsize=(8,5))
Region_Sales.plot(kind='bar',color='orange')
plt.title("Region wise Total Sales")
plt.xlabel("Region")
plt.ylabel("total Sales")
plt.xticks(rotation=0)
 plt.show()
```

```python
The above graph represents the total sales on the basis of Regions.thus,it is very claer that west region generates the highest total sales among all 
Regions ,indicating strong market performance and customer demand in this area.
The South region records the lowest total sales,suggesting potential growth opportunities or weaker market penetration.    
```

```python
df.columns
```

```python
Region_Profit=df.groupby("Region")["Profit"].sum().reset_index()
```

```python

sns.barplot(x="Region",y="Profit",data=Region_Profit,color="gold")
plt.title("Region vs Total Profit")
plt.xlabel("Region")
plt.ylabel("Total Profit")
plt.show()
```

```python
we can see that west is the most profit generating region while central region needs improvement and better strategy making.
```
